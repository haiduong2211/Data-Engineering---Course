```python
#remove larger number
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    def ll_append(self,data):
        new_node = Node(data)
        if self.headval == None:
            self.headval = new_node
            return    
        last_node = self.headval
        while last_node.nextval:
            last_node = last_node.nextval
        last_node.nextval = new_node

    def find_kval(self,k):
        cur_node = self.headval
        index = 0
        kval = None
        while cur_node:
            if index == k:
                kval = cur_node.dataval
                return kval
            index +=1
            cur_node = cur_node.nextval
        return kval

    def remove_larger(self,kval):
        cur_node = self.headval
        index = 0
        prev_node = None
        while index == 0:
            if kval < self.headval.dataval:
                self.headval = cur_node.nextval
                prev_node = cur_node
                cur_node = cur_node.nextval
                continue
            prev_node = cur_node
            cur_node = cur_node.nextval
            index +=1
                
        while cur_node:
            if kval < cur_node.dataval:
                prev_node.nextval = cur_node.dataval
                cur_node = cur_node.nextval
                continue

            prev_node = cur_node
            cur_node = cur_node.nextval
            index +=1

    def printlist(self):
        cur_node = self.headval
        while cur_node:
            print(cur_node.dataval, end = ' ')
            cur_node = cur_node.nextval
        return

llinked = LinkedList()
n = int(input())
for i in range(n):
    llinked.ll_append(int(input())) 


#linkedlist = [1,1,2,2,3,3]
#k = 2


k = int(input())
kval = llinked.find_kval(k)
llinked.remove_larger(kval)
llinked.printlist()
```
