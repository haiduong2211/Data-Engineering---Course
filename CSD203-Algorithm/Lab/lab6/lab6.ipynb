```python
#CodeLearn28 - Doublelinkedlist
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class doubly_linked_list:
    def __init__(self):
        self.head = None

    def push(self, NewVal):
        NewNode = Node(NewVal)
        NewNode.next = self.head
        if self.head is not None:
            self.head.prev = NewNode
        self.head = NewNode

    def append(self, NewVal):
        NewNode = Node(NewVal)
        NewNode.next = None
        if self.head is None:
            NewNode.prev = None
            self.head = NewNode
            return
        last = self.head
        while (last.next is not None):
            last = last.next
        last.next = NewNode
        NewNode.prev = last
        return
	
    def push(self, NewVal):
        NewNode = Node(NewVal)
        NewNode.next = self.head
        if self.head is not None:
            self.head.prev = NewNode
        self.head = NewNode
	
    def insert(self, prev_node, NewVal):
        NewNode = Node(NewVal)
        if prev_node is None:
            NewNode.next = self.head
            self.head = NewNode
            return
        NewNode.next = prev_node.next
        prev_node.next = NewNode
        NewNode.prev = prev_node
        if NewNode.next is not None:
            NewNode.next.prev = NewNode  

    def find_prev_node(self,k):
        if self.head is None:
            return None
        cur_node = self.head
        index = 0
        if k == 0:
            return None
        while index != k and cur_node.next is not None:
            cur_node = cur_node.next
            index +=1
        if index == k:
            return cur_node.prev
        if cur_node.next is None:
            return cur_node
        return cur_node.prev

    def listprint(self, node):
        while (node is not None):
            print(node.data,end = ' ')
            last = node
            node = node.next

llist = doubly_linked_list()
n = int(input())
for i in range(n):
    llist.append(int(input()))

k = int(input())
x = int(input())

llist.insert(llist.find_prev_node(k),x)
llist.listprint(llist.head)
```
